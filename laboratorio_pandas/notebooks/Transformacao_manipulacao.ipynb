# Transformação e manipulação de dados com Pandas

#### **Conhecendo os Dados**

```python
import pandas as pd
```

```python
caminho_arquivo = '../../data_extra/dados_hospedagem.json'
```

```python
dados_hospedagem = pd.read_json(caminho_arquivo)
dados_hospedagem.head()
```

#### **Compreendendo os Dados**

```python
dados_hospedagem = pd.json_normalize(dados_hospedagem['info_moveis'])
dados_hospedagem.head()
```

#### **Vamos entender o que cada coluna traz de informação** ✅

✅**avaliacao_geral**: refere-se à média de notas dadas para a avaliação da hospedagem no imóvel.

✅**experiencia_local**: descreve as experiências oferecidas durante a hospedagem no imóvel.

✅**max_hospedes**: informa a quantidade máxima de hóspedes que o local permite.

✅**descricao_local**: descreve o imóvel.

✅**descricao_vizinhanca**: descreve a vizinhança ao redor do imóvel.

✅**quantidade_banheiros**: informa a quantidade de banheiros disponíveis.

✅**quantidade_quartos**: informa a quantidade de quartos disponíveis.

✅**quantidade_camas**: informa a quantidade de camas disponíveis.

✅**modelo_cama**: informa o modelo de cama oferecido.

✅**comodidades**: informa as comodidades oferecidas pelo imóvel.

✅**taxa_deposito**: informa a taxa de depósito mínima para segurança de hospedagem.

✅**taxa_limpeza**: informa a taxa cobrada para o serviço de limpeza.

✅**preco**: refere-se ao preço base a ser cobrado pela diária no imóvel.
 Dis

```python
# criando uma lista de colunas
colunas = list(dados_hospedagem.columns)
colunas
```

```python
# desagrapando lista no dataframe usando explode
dados_hospedagem = dados_hospedagem.explode(colunas[3:])
dados_hospedagem.head()
```

```python
# Resetando o Index do Dataframe
dados_hospedagem.reset_index(inplace=True, drop=True)
dados_hospedagem.head()
```

```python
dados_hospedagem.info()
```

#### **Convertendo dados Numericos**

```python
import numpy as np
```

```python
colunas_numericas = ['quantidade_banheiros','quantidade_quartos','quantidade_camas','max_hospedes']
```

```python
dados_hospedagem[colunas_numericas] = dados_hospedagem[colunas_numericas].astype(np.int64)
dados_hospedagem.info()
```

```python
dados_hospedagem['avaliacao_geral'] = dados_hospedagem['avaliacao_geral'].astype(np.float64)
dados_hospedagem.info()
```

```python
lista_colunas = ['taxa_deposito',	'taxa_limpeza',	'preco']
```

```python
dados_hospedagem[lista_colunas] = dados_hospedagem[lista_colunas].replace(r'\$', '', regex=True).replace(r',', '', regex=True)
dados_hospedagem[lista_colunas] = dados_hospedagem[lista_colunas].astype(np.float64)
dados_hospedagem
```
