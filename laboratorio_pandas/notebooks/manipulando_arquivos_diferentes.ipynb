```python
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine, MetaData, Table, inspect, text
```

```python
engine = create_engine('sqlite:///:memory:')
```

```python
url_clientes  = 'https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/clientes_banco.csv'
```

```python
dados_clientes = pd.read_csv(url_clientes)
dados_clientes
```

```python
dados_clientes.to_sql('clientes', con=engine)
```

```python
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

```python
query = 'Select * from clientes where Categoria_de_renda = "Empregado"	'
```

```python
empregados = pd.read_sql_query(query, engine)
empregados
```

```python
empregados.to_sql('empregados', con=engine)
```

```python
pd.read_sql_table('empregados', engine)
```

```python
pd.read_sql_table('empregados', engine, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual'])
```

```python
# utilizar o TEXT
query = 'Delete from clientes where ID_Cliente=5008804'
with engine.connect() as con:
    con.execute(text(query))
```
